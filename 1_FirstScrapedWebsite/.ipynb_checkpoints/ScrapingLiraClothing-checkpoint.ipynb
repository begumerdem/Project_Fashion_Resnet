```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

import random
from random import randint
from time import sleep
```

```python
import urllib.request
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
url = "https://www.liraclothing.com/collections/mens-collection"
```

```python
response = requests.get(url)
response.status_code
```

```python
product_details = []
price = []
images_links =[]
new_image_list = []
pages = []

start_number = 1
max_number = 15

for number in range(start_number, max_number):
    response = requests.get('https://www.liraclothing.com/collections/mens-collection?page=' + str(number))
    soup = BeautifulSoup(response.content, "html.parser")
    soup.select("#shopify-section-collection > div > div:nth-child(3) > section > div > article:nth-child(1) > div")
    if number == 14:
        for i in range(6):
            product_details.append(soup.select("div.product--details")[i]("a")[0].div.get_text())
            price.append(soup.select("div.product--details")[i]("a")[1].span.get_text())
            image_list = soup.find_all("img",class_="lazyload")
            for i in range(len(image_list)):
                if len(image_list[i]['class']) == 1:
                    new_image_list.append(image_list[i])
                    for element in new_image_list:
                        images_links.append(element['data-src'])
                        images_links = list(dict.fromkeys(images_links))
    else:
        for i in range(16):
            product_details.append(soup.select("div.product--details")[i]("a")[0].div.get_text())
            price.append(soup.select("div.product--details")[i]("a")[1].span.get_text())
            image_list = soup.find_all("img",class_="lazyload")
            
            for i in range(len(image_list)):
                if len(image_list[i]['class']) == 1:
                    new_image_list.append(image_list[i])
                    for element in new_image_list:
                        images_links.append(element['data-src'])
                        images_links = list(dict.fromkeys(images_links))
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " second(s).")
    sleep(wait_time)
    # store response into "pages" list
    pages.append(response)
```

```python
for i in range(len(images_links)):
    images_links[i] = re.sub("{width}","180",images_links[i])
images_links
```

```python
product_prices=[]
for i in price:
    x = re.sub("\n","",i)
    y = re.sub("\n","",x)
    product_prices.append(x)
```

```python
import pandas as pd

products = pd.DataFrame({"product_details":product_details,
                       "price":product_prices, "image_link":images_links})

products.to_csv("LiraClothing.csv",index=False)
products
```
